```python
import pandas as pd
df_april_may = pd.read_csv("Classes April-May 2018.csv")
df_april_may.head()
```

```python
df_april_may.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_april_may.head()
```

```python
df_april_may["BookingEndDate"] = pd.to_datetime(df_april_may["BookingEndDate"], errors='coerce')
df_april_may["BookingStartTime"] = pd.to_datetime(df_april_may["BookingStartTime"], format='%H:%M:%S', errors='coerce').dt.time
df_april_may.head()
```

```python
df_april_may["PriceINR"].fillna(df_april_may["PriceINR"].median(), inplace=True)
df_april_may.head()
```

```python
df_april_may["DayOfWeek"] = df_april_may["BookingEndDate"].dt.day_name()
df_april_may["OccupancyRate"] = df_april_may["NumberBooked"] / df_april_may["MaxBookees"]
df_april_may.head()
```

```python
df_april_may.to_csv("Classes_April_to_May_2018.csv", index=False)
```

```python
import pandas as pd
df_june = pd.read_csv("Classes June 2018.csv")
df_june.head()
```

```python
df_june.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_june.head()
```

```python
df_june["BookingEndDate"] = pd.to_datetime(df_june["BookingEndDate"], errors='coerce')
df_june["BookingStartTime"] = pd.to_datetime(df_june["BookingStartTime"], format="%H:%M:%S", errors='coerce').dt.time
df_june.head()
```

```python
df_june["DayOfWeek"] = df_june["BookingEndDate"].dt.day_name()
df_june["OccupancyRate"] = df_june["NumberBooked"] / df_june["MaxBookees"]
df_june.head()
```

```python
df_june.to_csv("Classes_June_2018.csv", index=False)
```

```python
df_merged = pd.merge(df_april_may, df_june, on=["ActivitySiteID", "ActivityDescription", "BookingEndDate"],suffixes=('_AprMay', '_June'),how='outer')
df_merged = pd.concat([df_april_may, df_june], ignore_index=True)
df_merged
```

```python
df_merged["BookingEndDate"] = pd.to_datetime(df_merged["BookingEndDate"], errors='coerce')
df_merged["BookingStartTime"] = pd.to_datetime(df_merged["BookingStartTime"], errors='coerce').dt.time
df_merged["PriceINR"] = pd.to_numeric(df_merged["PriceINR"], errors='coerce')

df_merged["PriceINR"].fillna(df_merged["PriceINR"].median(), inplace=True)
df_merged
```

```python
quality_report = {}
null_counts = df_merged.isnull().sum()
quality_report['Missing Values'] = null_counts[null_counts > 0]
duplicates = df_merged.duplicated().sum()
quality_report['Duplicate Rows'] = duplicates
invalid_prices = df_merged[df_merged["PriceINR"] <= 0].shape[0]
quality_report['Invalid Prices (<= 0)'] = invalid_prices
invalid_occupancy = df_merged[df_merged["OccupancyRate"] > 1].shape[0]
quality_report['Invalid Occupancy (>1)'] = invalid_occupancy
```

```python
print("=== DATA QUALITY REPORT ===")
for issue, value in quality_report.items():
    print(f"{issue}: {value}")
```

```python
df_merged.to_csv("df_merged2018.csv", index=False)
```

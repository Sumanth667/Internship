```python
import pandas as pd
df_april_may = pd.read_csv("Classes April-May 2018.csv")
df_april_may.head()
```

```python
df_april_may.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_april_may.head()
```

```python
df_april_may["BookingEndDate"] = pd.to_datetime(df_april_may["BookingEndDate"], errors='coerce')
df_april_may["BookingStartTime"] = pd.to_datetime(df_april_may["BookingStartTime"], format='%H:%M:%S', errors='coerce').dt.time
df_april_may.head()
```

```python
df_april_may["PriceINR"].fillna(df_april_may["PriceINR"].median(), inplace=True)
df_april_may.head()
```

```python
df_april_may["DayOfWeek"] = df_april_may["BookingEndDate"].dt.day_name()
df_april_may["OccupancyRate"] = df_april_may["NumberBooked"] / df_april_may["MaxBookees"]
df_april_may.head()
```

```python
df_april_may.to_csv("Classes_April_to_May_2018.csv", index=False)
```

```python
import pandas as pd
df_june = pd.read_csv("Classes June 2018.csv")
df_june.head()
```

```python
df_june.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_june.head()
```

```python
df_june["BookingEndDate"] = pd.to_datetime(df_june["BookingEndDate"], errors='coerce')
df_june["BookingStartTime"] = pd.to_datetime(df_june["BookingStartTime"], format="%H:%M:%S", errors='coerce').dt.time
df_june.head()
```

```python
df_june["DayOfWeek"] = df_june["BookingEndDate"].dt.day_name()
df_june["OccupancyRate"] = df_june["NumberBooked"] / df_june["MaxBookees"]
df_june.head()
```

```python
df_june.to_csv("Classes_June_2018.csv", index=False)
```

```python
df_combined = pd.concat([df_april_may, df_june], ignore_index=True)
print("Combined Data Shape:", df_combined.shape)
df_combined.head()
```

```python
print("Missing values after merging:")
print(df_combined.isnull().sum())
df_combined["PriceINR"].fillna(df_combined["PriceINR"].median(), inplace=True)
```

```python
duplicates = df_combined.duplicated().sum()
print("Duplicate records:", duplicates)
```

```python
invalid_occupancy = df_combined[df_combined["OccupancyRate"] > 1]
print("Invalid Occupancy Rate Records:", len(invalid_occupancy))
```

```python
df_combined["BookingEndDate"] = pd.to_datetime(df_combined["BookingEndDate"], errors='coerce')
quality_report = {
    "Total Records": len(df_combined),
    "Missing Values": df_combined.isnull().sum().to_dict(),
    "Duplicate Records": duplicates,
    "Invalid Occupancy Rate (>1)": len(invalid_occupancy),
    "Unique Activities": df_combined["ActivityDescription"].nunique(),
    "Date Range": (df_combined["BookingEndDate"].min(), df_combined["BookingEndDate"].max())
}
import pprint
pprint.pprint(quality_report)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
df_model = df_combined[(df_combined["PriceINR"] > 0) & (df_combined["NumberBooked"] > 0)].copy()
df_model["log_price"] = np.log(df_model["PriceINR"])
df_model["log_demand"] = np.log(df_model["NumberBooked"])
X = df_model[["log_price"]]
y = df_model["log_demand"]
```

```python
model = LinearRegression()
model.fit(X, y)

elasticity = model.coef_[0]
intercept = model.intercept_

print(f"Estimated Price Elasticity of Demand: {elasticity:.4f}")
```

```python
y_pred = model.predict(X)
residuals = y - y_pred
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f"R: {r2:.4f}")
print(f"Mean Squared Error: {mse:.4f}")

plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Predicted Log(Demand)")
plt.ylabel("Residuals")
plt.tight_layout()
plt.show()
```

```python
df_timeseries = df_combined.groupby("BookingEndDate")["NumberBooked"].sum().reset_index()

# Set the date as index
df_timeseries = df_timeseries.set_index("BookingEndDate")
df_timeseries = df_timeseries.asfreq("D")  # Set daily frequency
df_timeseries = df_timeseries.fillna(0)    # Fill missing days with 0 bookings

df_timeseries.head()
```

```python
from prophet import Prophet
df_prophet = df_timeseries.reset_index().rename(columns={"BookingEndDate": "ds", "NumberBooked": "y"})
model = Prophet(daily_seasonality=True)
model.fit(df_prophet)

future = model.make_future_dataframe(periods=15) 
forecast = model.predict(future)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
df_merge = pd.merge(df_prophet, forecast[['ds', 'yhat']], on='ds')
mse = mean_squared_error(df_merge['y'], df_merge['yhat'])
mae = mean_absolute_error(df_merge['y'], df_merge['yhat'])

print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
model.plot(forecast)
plt.title("Demand Forecast for Fitness Classes")
plt.xlabel("Date")
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.show()

model.plot_components(forecast)
plt.tight_layout()
plt.show()
```
